```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("results.csv")
```

```python
df.head()
```

```python
r2_df = df.melt(id_vars="DATASET",value_vars=["NG_R2","RD_R2"])
r2_df.columns = ["Dataset","Method","R**2"]
```

```python
r2_df.head()
```

```python
ax = sns.boxplot(x="Dataset",y="R**2",hue="Method",data=r2_df,order=sorted(r2_df.Dataset.unique()))
sns.set(rc={'figure.figsize': (15, 12)})
sns.set(font_scale=1.5)
sns.set_style('whitegrid')
ax.tick_params(axis='x', rotation=90)
```

```python
rms_df = df.melt(id_vars="DATASET",value_vars=["NG_RMSE","RD_RMSE"])
rms_df.columns = ["Dataset","Method","RMSE"]
```

```python
rms_df.head()
```

```python
ax = sns.boxplot(x="Dataset",y="RMSE",hue="Method",data=rms_df,order=sorted(rms_df.Dataset.unique()))
sns.set(rc={'figure.figsize': (15, 12)})
sns.set(font_scale=1.5)
sns.set_style('whitegrid')
ax.tick_params(axis='x', rotation=90)
```
